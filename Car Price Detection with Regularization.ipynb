```python
# Sorting Selling price of car by descending order
df=df.sort_values(by='Selling_Price',ascending=False).reset_index()
```

```python
df.head()
```

```python
# Removing outliers and and resetting index
df=df[3:].reset_index()
```

```python
df.head()
```

```python
# Dropping extra index'Level 0'
df.drop(['level_0','index'],axis=1)
```

```python
# Splitting data into X and Y variable
```

```python
X= df.drop(columns='Selling_Price')
y=df['Selling_Price']
```

```python
#Using oneHot encoder for column transformation from 'string' to 'int'
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_cols=['Car_Name','Fuel_Type','Seller_Type','Transmission']
transformer= ColumnTransformer([('OHE',OneHotEncoder(),categorical_cols)],
                            remainder='passthrough')
X=transformer.fit_transform(X)
```

```python
# Spliting the data in train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size= 0.3)
```

```python
X_train.shape , X_test.shape
```

```python
# Appling multiple Linear regression on train and test data
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
# predicting the selling price
y_predictions=model.predict(X_test)
```

```python
y_predictions
```

```python
# Checking metrics like r2 ,r2 score,mae
from sklearn.metrics import mean_absolute_error,r2_score
MAE= mean_absolute_error(y_test,y_predictions)
R_squared=r2_score(y_test,y_predictions)

n=len(y_test)
p=X_test.shape[1]
Adjusted_R_squared=1-(1-R_squared)*(n-1)/(n-p-1)
```

```python
r2_score(y_test,y_predictions)
```
